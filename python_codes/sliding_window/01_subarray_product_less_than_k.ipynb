Given an array of integers nums and an integer k, return the number of contiguous subarrays where the product of all the elements in the subarray is strictly less than k.

 

Example 1:

Input: nums = [10,5,2,6], k = 100
Output: 8
Explanation: The 8 subarrays that have product less than 100 are:
[10], [5], [2], [6], [10, 5], [5, 2], [2, 6], [5, 2, 6]
Note that [10, 5, 2] is not included as the product of 100 is not strictly less than k.
Example 2:

Input: nums = [1,2,3], k = 0
Output: 0
 

Constraints:

1 <= nums.length <= 3 * 104
1 <= nums[i] <= 1000
0 <= k <= 106

```python
tests = []

test1 = {
    "inputs" : {
        "array" : [10,5,2,6],
        "target" : 100
    },
    "output": 8
}

test2 = {
    "inputs" : {
        "array" : [1,2,3],
        "target" : 0
    },
    "output": 0
}

test3= {
    "inputs" : {
        "array" : [2,5,6,10],
        "target" : 100
    },
    "output": 8
}


tests.append(test1)
tests.append(test2)
tests.append(test3)
print(tests)
```

```python
def subarray_product_less_than_k(array, target):
    count = 0
    array.sort()
    i = j = 0
    temp_subarray_product = array[j]
    while j<len(array):
        if temp_subarray_product<target:
            j = j+1
        else:
            count+= (2**(j-i))-1
            i = i+1
        temp_subarray_product = temp_subarray_product*array[j]
    
    print(f'count is {count}')
    return count
```

```python
for test in tests:
    fn_output = subarray_product_less_than_k(**test["inputs"])
    print(f'{fn_output == test["output"]} \n')
```

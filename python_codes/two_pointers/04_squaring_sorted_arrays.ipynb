https://leetcode.com/problems/squares-of-a-sorted-array/description/

Given an integer array nums sorted in non-decreasing order, return an array of the squares of each number sorted in non-decreasing order.

 
Example 1:

Input: nums = [-4,-1,0,3,10]
Output: [0,1,9,16,100]
Explanation: After squaring, the array becomes [16,1,0,9,100].
After sorting, it becomes [0,1,9,16,100].
Example 2:

Input: nums = [-7,-3,2,3,11]
Output: [4,9,9,49,121]
 

Constraints:

1 <= nums.length <= 104
-104 <= nums[i] <= 104
nums is sorted in non-decreasing order.
 

Follow up: Squaring each element and sorting the new array is very trivial, could you find an O(n) solution using a different approach?

```python
tests = []

test1 = {
    "inputs": [-2,-1,0,3,4],
    "output": [0,1,4,9,16]
}


test2 = {
    "inputs": [-4,-1,0,3,10],
    "output": [0,1,9,16,100]
}

test3 = {
    "inputs": [-7,-3,2,3,11],
    "output": [4,9,9,49,121]
}

test3 = {
    "inputs": [-7,-1,2,2,3,11],
    "output": [1,4,4,9,49,121]
}

test4 = {
    "inputs": [-7,-3,2,2,3,11],
    "output": [4,4,9,9,49,121]
}


test5 = {
    "inputs": [-7,-3,-2,2,2,3,11],
    "output": [4,4,4,9,9,49,121]
}

tests.append(test1)
tests.append(test2)
tests.append(test3)
tests.append(test4)
tests.append(test5)

print(tests)
```

M1: Idea is that we shall hold the pointer i always at index 0 and and we shall iterate using reverse order using pointer j.

Compare the values of index i and index and swap. however this algo is failing for some test cases.

```python
def squaring_sorted_arrayes(nums):
    i = 0
    j = len(nums)-1
    while j>=0:
        if abs(nums[i])> abs(nums[j]):
            temp = nums[i]
            nums[i] = nums[j]
            nums[j]= temp*temp
        else:
            nums[j] = nums[j]*nums[j] 
        j-=1
    return nums
```

```python
for test in tests:
    print(f'input is     {test["inputs"]}')
    fn_output = squaring_sorted_arrayes(test["inputs"])
    print(f'expected output is {test["output"]} \nand actual output is {fn_output}')
    print(f'{fn_output == test["output"]} \n')
```

M2: We shall use two moving pointers in opposite direction and use additional variable to store the array. however this will add the space complexity of O(n)

Note: That in 2 pointer algorithms, both pointers ALWAYS move. Hence M1 is not a valid two pointer algo.

```python
tests = []

test1 = {
    "inputs": [-2,-1,0,3,4],
    "output": [0,1,4,9,16]
}


test2 = {
    "inputs": [-4,-1,0,3,10],
    "output": [0,1,9,16,100]
}

test3 = {
    "inputs": [-7,-3,2,3,11],
    "output": [4,9,9,49,121]
}

test3 = {
    "inputs": [-7,-1,2,2,3,11],
    "output": [1,4,4,9,49,121]
}

test4 = {
    "inputs": [-7,-3,2,2,3,11],
    "output": [4,4,9,9,49,121]
}


test5 = {
    "inputs": [-7,-3,-2,2,2,3,11],
    "output": [4,4,4,9,9,49,121]
}

tests.append(test1)
tests.append(test2)
tests.append(test3)
tests.append(test4)
tests.append(test5)

print(tests)
```

```python
def squaring_sorted_arrayes_2_pointers(nums):
    i=0
    j=len(nums)-1
    k = len(nums)-1
    output= [0] * len(nums)
    while i<=j:
        numisq = nums[i]*nums[i]
        numjsq = nums[j] * nums[j]
        if numisq < numjsq:
            output[k] = numjsq
            j-=1
            k-=1
        elif numisq > numjsq:
            output[k] = numisq
            i+=1
            k-=1
        elif numisq == numjsq:
            output[k] = numisq
            i+=1
            k-=1
    return output
```

```python
for test in tests:
    print(f'input is     {test["inputs"]}')
    fn_output = squaring_sorted_arrayes_2_pointers(test["inputs"])
    print(f'expected output is {test["output"]} \nand actual output is {fn_output}')
    print(f'{fn_output == test["output"]} \n')
```

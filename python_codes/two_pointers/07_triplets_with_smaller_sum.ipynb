Given an array arr[] of distinct integers of size n and a value sum, the task is to find the count of triplets (i, j, k), having (i<j<k) with the sum of (arr[i] + arr[j] + arr[k]) smaller than the given value sum.


Examples :


Input: n = 4, sum = 2, arr[] = {-2, 0, 1, 3}
Output:  2
Explanation: Below are triplets with sum less than 2 (-2, 0, 1) and (-2, 0, 3). 
Input: n = 5, sum = 12, arr[] = {5, 1, 3, 4, 7}
Output: 4
Explanation: Below are triplets with sum less than 12 (1, 3, 4), (1, 3, 5), (1, 3, 7) and (1, 4, 5).
Expected Time Complexity: O(N2).
Expected Auxiliary Space: O(1).


Constraints:
3 ≤ N ≤ 103
-103 ≤ arr[i] ≤ 103

```python
tests =[]


test1 = {
    "inputs": {
        "size" : 4,
        "array" : [-2, 0, 1, 3],
        "sum" : 2
    },
    "output" : 2
}


test2 = {
    "inputs": {
        "size" : 5,
        "array" : [5, 1, 3, 4, 7],
        "sum" : 12
    },
    "output" : 4
}

tests.append(test1)
tests.append(test2)
```

```python
def triplets_with_smaller_sum(size,array,sum):
    array.sort()
    triplet_count = 0
    print(f'inputs are {size, array, sum}')
    for i in range(size-2):
        j=i+1
        k=size-1
        while j<k:
            print(f'indexes are {i, j, k} and values are {array[i], array[j], array[k]}')
            current_sum = array[i]+array[j]+array[k]
            if current_sum < sum:
                triplet_count+= k-j
                print(f'triplet count is {triplet_count}')
                j+=1
            else:
                k-=1
    return triplet_count
```

```python
for test in tests:
    fn_output = triplets_with_smaller_sum(**test["inputs"])
    print(f'{fn_output == test["output"]} \n')
```

https://leetcode.com/problems/3sum/

Given an integer array nums, return all the triplets [nums[i], nums[j], nums[k]] such that i != j, i != k, and j != k, and nums[i] + nums[j] + nums[k] == 0.

Notice that the solution set must not contain duplicate triplets.

 

Example 1:

Input: nums = [-1,0,1,2,-1,-4]
Output: [[-1,-1,2],[-1,0,1]]
Explanation: 
nums[0] + nums[1] + nums[2] = (-1) + 0 + 1 = 0.
nums[1] + nums[2] + nums[4] = 0 + 1 + (-1) = 0.
nums[0] + nums[3] + nums[4] = (-1) + 2 + (-1) = 0.
The distinct triplets are [-1,0,1] and [-1,-1,2].
Notice that the order of the output and the order of the triplets does not matter.
Example 2:

Input: nums = [0,1,1]
Output: []
Explanation: The only possible triplet does not sum up to 0.
Example 3:

Input: nums = [0,0,0]
Output: [[0,0,0]]
Explanation: The only possible triplet sums up to 0.
 

Constraints:

3 <= nums.length <= 3000
-105 <= nums[i] <= 105

```python
tests = []

test1 = {
    "inputs": [-1,0,1,2,-1,-4],
    "output": [[-1,-1,2],[-1,0,1]]
}

test2 = {
    "inputs": [0,1,1],
    "output": []
}

test3 = {
    "inputs": [0,0,0],
    "output": [[0,0,0]]
}

test4 = {
    "inputs": [0,0,0,0],
    "output": [[0,0,0]]
}

test5 = {
    "inputs": [-100,-70,-60,110,120,130,160],
    "output": [[-100,-60,160],[-70,-60,130]]
}


tests.append(test1)
tests.append(test2)
tests.append(test3)
tests.append(test4)
tests.append(test5)
```

```python
def threeSum(nums):
    """
    :type nums: List[int]
    :rtype: List[List[int]]
    """
    i = 0 
    output = []
    nums.sort()
    print(f'sorted numbers is {nums}')
    for i in range(len(nums)-2):
        j = i+1
        k = len(nums)-1
        while j<k:
            print(f'i is {i}, j is {j}, k is {k}')
            if (nums[i]+nums[j]+nums[k] == 0):
                triplet = [nums[i],nums[j],nums[k]]
                if triplet not in output:
                    output.append(triplet)
                j+=1
            elif (nums[i]+nums[j]+nums[k] < 0):
                j+=1 
            else:
                 k-=1
    return output
```

```python
for test in tests:
    print(f'input is     {test["inputs"]}')
    fn_output = threeSum(test["inputs"])
    print(f'expected output is {test["output"]} \nand actual output is {fn_output}')
    print(f'{fn_output == test["output"]} \n')
```

As per leetcode submission of this code. Complexity is O(n**2)
And space complexity is O(n)

*However we are not solving the duplicate triplets and we are using python in built mechanisms

Cleaner solution solution from  leetcode problems using continue to skip the duplicates.
https://leetcode.com/problems/3sum/solutions/7008096/3sum-in-just-2-loops-beat-100-with-this-joid0/

```python
def three_sum_without_duplicates(nums):
    nums.sort()
    output = []
    for i in range(len(nums)-2):
        if i>0 and (nums[i] == nums[i-1]): continue
        j=i+1
        k=len(nums)-1
        while j<k:
            total = nums[i]+nums[j]+nums[k]
            if total == 0:
                output.append([nums[i],nums[j],nums[k]])                
                while j<k and (nums[j] == nums[j+1]): j+=1
                while j<k and (nums[k] == nums[k-1]): k-=1   
                j+=1
                k-=1
            elif total < 0:
                j+=1
            elif total > 0:
                k-=1
    return output

```

```python
for test in tests:
    print(f'input is     {test["inputs"]}')
    fn_output = three_sum_without_duplicates(test["inputs"])
    print(f'expected output is {test["output"]} \nand actual output is {fn_output}')
    print(f'{fn_output == test["output"]} \n')
```

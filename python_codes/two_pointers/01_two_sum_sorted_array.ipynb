Given a 1-indexed array of integers numbers that is already sorted in non-decreasing order, find two numbers such that they add up to a specific target number. Let these two numbers be numbers[index1] and numbers[index2] where 1 <= index1 < index2 <= numbers.length.

Return the indices of the two numbers, index1 and index2, added by one as an integer array [index1, index2] of length 2.

The tests are generated such that there is exactly one solution. You may not use the same element twice.

Your solution must use only constant extra space.

```python
tests = []

test1 = {
    "input" : {
        "numbers" : [1, 5 ,6 , 8],
        "target" : 7
    },
    "output" : [1,3],
}

test2 = {
    "input" : {
        "numbers" : [1, 5 ,6 , 8],
        "target" : 9
    },
    "output" : [1,4],
}


test3 = {
    "input" : {
        "numbers" : [1, 5 ,6 , 7, 8],
        "target" : 9
    },
    "output" : [1,5],
}

test4 = {
    "input" : {
        "numbers" : [1, 5 ,6 , 7, 8],
        "target" : 11
    },
    "output" : [2,3],
}


test5 = {
    "input" : {
        "numbers" : [-1, 3 ,6 , 7, 8],
        "target" : 2
    },
    "output" : [1,2],
}

tests.append(test1)
tests.append(test2)
tests.append(test3)
tests.append(test4)
tests.append(test5)
```

M1: #Brute force

```python
#Brute force

def two_sum_sorted_array(numbers, target):
    for i in range(len(numbers)):
        for j in range(len(numbers)):
            if i!=j:
                if numbers[i] + numbers [j] == target:
                    return [i+1, j+1]
```

```python
for test in tests:
    assert two_sum_sorted_array(**test["input"]) == test["output"]
```

Complexity : O(n^2)

M3: Two pointer

```python
def two_pointer(numbers,target):
    i = 0
    j = len(numbers) -1
    while i<j:
        if numbers[i] + numbers[j] == target:
            return [i+1, j+1]
        elif numbers[i] + numbers[j] < target:
            i = i+1
        elif numbers[i] + numbers[j] > target:
            j = j-1
```

```python
for test in tests:
    print(f'input is              {test["input"]}')
    fn_output = two_pointer(**test["input"])     
    print(f'expected output is    {test["output"]} and \nfn_output is          {fn_output} \n')
    assert fn_output == test["output"]
```
